```python
normalization_df=pd.DataFrame({'x':np.random.randint(-100,100,1000),
                               'y':np.random.randint(-80,80,1000),
                               'z':np.random.randint(-150,150,1000)})
```

```python
normalization_df
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(normalization_df.x,normalization_df.y,normalization_df.z)
plt.figure()
```

```python
from sklearn.preprocessing import Normalizer
```

```python
normal = Normalizer()
```

```python
normalization_df = normal.fit_transform(normalization_df)
```

```python
normalization_df
```

```python
norm_df = pd.DataFrame(normalization_df,columns=['x1','x2','x3'])
```

```python
norm_df
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(norm_df.x1,norm_df.x1,norm_df.x3)
plt.figure()
```

#  Missing values

```python
import numpy as np
```

```python
import pandas as pd
```

```python
data = {'First score':[100,90,np.nan,95],'Second score':[30,np.nan,56,np.nan],'Third score':[np.nan,40,np.nan,np.nan]}
```

```python
scores = pd.DataFrame(data)
```

```python
scores
```

```python
scores.isnull()
```

```python
scores.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy='median')
```

```python
imputer.fit(scores)
```

```python
imputer.transform(scores)
```
