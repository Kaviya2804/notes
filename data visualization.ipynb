# matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies = ['MI2','Fast&Furious','HarryPotter','Ironman','Spiderman',]
percentage = [65,75,90,66,67,]
```

```python
plt.bar(movies,percentage,color="green",edgecolor="red")
```

```python
plt.scatter(movies,percentage,color="red")
```

```python
plt.plot(movies,percentage,linestyle='solid')
```

```python
plt.pie(percentage,labels=movies)
```

# Sine wave

```python
import numpy as np
```

```python
x = np.arange(-10.0,10,0.02)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
plt.show()
```

# Relationship plot

```python
sns.relplot(x='total_bill',y='tip',data=tip,col="time",
            hue="day",style="day",kind="scatter")
```

 # Box plot

```python
sns.catplot(x='day', y='total_bill', kind='box', data=tip)
```

# Regression plot

```python
sns.regplot(x='total_bill', y='tip', data=tip)
```
